```
import matplotlib.pyplot as plt
import numpy as np

from matplotlib.axes._axes import _log as M_A_L
M_A_L.setLevel('ERROR')

t = np.arange(0.0, 2.0, 0.01)
Sin_1 = np.sin(np.pi * t)
Sin_2 = np.cos(2 * np.pi * t)
Sin_3 = np.sin(3 * np.pi * t)
Sin_4 = np.sin(4 * np.pi * t)

Figure = plt.figure(figsize = (10, 10))

Axes_1 = Figure.add_subplot(2, 2, 1)
Axes_1.plot(t, Sin_1, color = 'Black', linewidth = 4.0)
Axes_1.set(xlabel = 'Time (с)', ylabel = 'Voltage (mВ)', title='Sin')
Axes_1.grid()

Axes_2 = Figure.add_subplot(2, 2, 2)
Axes_2.plot(t, Sin_2, color = 'Yellow',linewidth=4.0)
Axes_2.set(xlabel = 'Time (с)', ylabel = 'Voltage (mВ)', title='Sin')
Axes_2.grid()

Axes_3 = Figure.add_subplot(2, 2, 3)
Axes_3.plot(t, Sin_3, c = 'Blue',linewidth=4.0)
Axes_3.set(xlabel = 'Time', ylabel = 'Voltage (mВ)', title='Sin')
Axes_3.grid()

Axes_4 = Figure.add_subplot(2, 2, 4)
Axes_4.plot(t, Sin_4, color = 'Green', linewidth = 4.0)
Axes_4.set(xlabel = 'Time (с)', ylabel = 'Voltage (mВ)', title = 'Sin')
Axes_4.grid()

plt.show()
```
