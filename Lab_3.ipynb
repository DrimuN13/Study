```
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```
img_half = cv2.imread('Lab3_half100.png', cv2.IMREAD_GRAYSCALE)
img_chess = cv2.imread('Lab3_chess100.png', cv2.IMREAD_GRAYSCALE)
```

```
hist_half = cv2.calcHist([img_half], [0], None, [256], [0, 256])
hist_chess = cv2.calcHist([img_chess], [0], None, [256], [0, 256])

plt.figure(figsize=(20, 10))
plt.subplot(121)
plt.imshow(img_half, cmap = 'gray');
plt.subplot(122)
plt.plot(hist_half);
plt.xlim([0, 255]);
plt.grid();
plt.show();

plt.figure(figsize=(20,10))
plt.subplot(121)
plt.imshow(img_chess, cmap = 'gray');
plt.subplot(122)
plt.plot(hist_chess);
plt.xlim([0, 255]);
plt.grid();
plt.show();

plt.show();
```

```
half_blur_3 = cv2.cv2.blur(img_half, (3, 3))
half_blur_5 = cv2.cv2.blur(img_half, (5, 5))
half_blur_9 = cv2.cv2.blur(img_half, (9, 9))
half_blur_15 = cv2.cv2.blur(img_half, (15, 15))

chess_blur_3 = cv2.blur(img_chess, (3, 3))
chess_blur_5 = cv2.blur(img_chess, (5, 5))
chess_blur_9 = cv2.blur(img_chess, (9, 9))
chess_blur_15 = cv2.blur(img_chess, (15, 15))

half_blur_3_hist = cv2.calcHist([half_blur3], [0], None, [256], [0, 256])
half_blur_5_hist = cv2.calcHist([half_blur5], [0], None, [256], [0, 256])
half_blur_9_hist = cv2.calcHist([half_blur9], [0], None, [256], [0, 256])
half_blur_15_hist = cv2.calcHist([half_blur15], [0], None, [256], [0, 256])

chess_blur_3_hist = cv2.calcHist([chess_blur3], [0], None, [256], [0, 256])
chess_blur_5_hist = cv2.calcHist([chess_blur5], [0], None, [256], [0, 256])
chess_blur_9_hist = cv2.calcHist([chess_blur9], [0], None, [256], [0, 256])
chess_blur_15_hist = cv2.calcHist([chess_blur15], [0], None, [256], [0, 256])

plt.figure(figsize=(20, 10))
plt.subplot(121)
plt.imshow(half_blur_3, cmap = 'gray');
plt.subplot(122)
plt.plot(half_blur_3_hist);
plt.xlim([0, 255]);
plt.grid();
plt.show();

plt.figure(figsize=(20, 10))
plt.subplot(121)
plt.imshow(half_blur_5, cmap = 'gray');
plt.subplot(122)
plt.plot(half_blur_5_hist);
plt.xlim([0, 255]);
plt.grid();
plt.show();

plt.figure(figsize=(20, 10))
plt.subplot(121)
plt.imshow(half_blur_9, cmap = 'gray');
plt.subplot(122)
plt.plot(half_blur_9_hist);
plt.xlim([0, 255]);
plt.grid();
plt.show();

plt.figure(figsize=(20, 10))
plt.subplot(121)
plt.imshow(half_blur_15, cmap = 'gray');
plt.subplot(122)
plt.plot(half_blur_15_hist);
plt.xlim([0, 255]);
plt.grid();
plt.show();

plt.figure(figsize=(20, 10))
plt.subplot(121)
plt.imshow(chess_blur_3, cmap = 'gray');
plt.subplot(122)
plt.plot(chess_blur_3_hist);
plt.xlim([0, 255]);
plt.grid();
plt.show();

plt.figure(figsize=(20, 10))
plt.subplot(121)
plt.imshow(chess_blur_5, cmap = 'gray');
plt.subplot(122)
plt.plot(chess_blur_5_hist);
plt.xlim([0, 255]);
plt.grid();
plt.show();

plt.figure(figsize=(20, 10))
plt.subplot(121)
plt.imshow(chess_blur_9, cmap = 'gray');
plt.subplot(122)
plt.plot(chess_blur_9_hist);
plt.xlim([0, 255]);
plt.grid();
plt.show();

plt.figure(figsize=(20, 10))
plt.subplot(121)
plt.imshow(chess_blur_15, cmap = 'gray');
plt.subplot(122)
plt.plot(chess_blur_15_hist);
plt.xlim([0, 255]);
plt.grid();
plt.show();
```
