```
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```
img = cv2.imread("1_8.png", cv2.IMREAD_GRAYSCALE)
plt.figure(figsize = (5, 5))
plt.imshow(img)
plt.title('Normal')
plt.show()

bin_img = img.copy()
threshold = 10
max_val = 255
min_val = 0

for row in range(img.shape[0]):
    for col in range(img.shape[1]):
        if (img[row][col]) > threshold:
            bin_img[row][col] = max_val
        else:
            bin_img[row][col] = min_val

plt.figure(figsize = (5, 5))
plt.imshow(bin_img)
plt.title('Binary')
plt.show()

#plot erosin
kernel = np.ones((2, 2), np.uint8)
erosion = cv2.erode(bin_img, kernel, iterations = 1)
plt.figure(figsize = (5, 5))
plt.imshow(erosion, cmap = 'gray')
plt.title('Erosion')
plt.show()

#plot dilation
dilation = cv2.dilate(bin_img, kernel, iterations = 1)
plt.figure(figsize = (5, 5))
plt.imshow(dilation, cmap = 'gray')
plt.title('Dilation')
plt.show()
```

```
img = cv2.imread("street.jpg", cv2.IMREAD_GRAYSCALE)
plt.figure(figsize=(5,5))
plt.imshow(img) 
plt.title('Normal')
plt.show()

bin_img = img.copy()
threshold = 170
max_val = 255
min_val = 0

for row in range(img.shape[0]):
    for col in range(img.shape[1]):
        if (img[row][col]) > threshold:
            bin_img[row][col] = max_val
        else:
            bin_img[row][col] = min_val

plt.figure(figsize = (5, 5))
plt.imshow(bin_img)
plt.title('Binary')
plt.show()

kernel = np.ones((2, 2), np.uint8)
erosion = cv2.erode(bin_img, kernel, iterations = 1)
plt.figure(figsize = (5, 5))
plt.imshow(erosion, cmap = 'gray')
plt.title('Erosion')
plt.show()

opening = cv2.morphologyEx(bin_img,  cv2.MORPH_OPEN, kernel,iterations = 1)
plt.figure(figsize = (5, 5))
plt.imshow(opening, cmap = 'gray')
plt.title('Opening')
plt.show()
```
