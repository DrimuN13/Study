```
import cv2
import matplotlib.pyplot as plt
import numpy as np

img_half = cv2.imread("Half.png")
img_chess = cv2.imread("Chess.png")

hist_half = cv2.calcHist([img_half], [0], None, [256], [0, 256])
hist_chess = cv2.calcHist([img_chess], [0], None, [256], [0, 256])

plt.figure(figsize = (10, 5))
plt.subplot(121)
plt.imshow(img_half, cmap = "gray")
plt.subplot(122)
plt.plot(hist_half); plt.xlim([0, 255]); plt.ylim([0, 5100]); plt.grid()
plt.show()

plt.figure(figsize = (10, 5))
plt.subplot(121)
plt.imshow(img_chess, cmap = "gray")
plt.subplot(122)
plt.plot(hist_chess); plt.xlim([0, 255]); plt.ylim([0, 5100]); plt.grid()
plt.show()
```

```
r = range(0, 64)
chess_board = np.zeros((8, 8), dtype = 'uint8')
for i in range(0, 8):
    for j in range (0, 8):
        if r[i + j] % 2 != 0:
            chess_board[i, j] = 1

print("'0' - white color\n'1' - black color\n")
print(chess_board)
```
